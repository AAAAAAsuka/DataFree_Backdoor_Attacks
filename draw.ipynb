```python
def draw(exp, model, dataset, mode, xlabel, ylabel="ACC/ASR"):
    plt.figure()
    if exp is None:
        data = np.load(f'results/{mode}_{model}_{dataset}.npy')
    else:
        data = np.load(f'results/{mode}_{exp}_{model}_{dataset}.npy')
    x, ACC, ASR = data[:,0], data[:,1], data[:,2]
    width_c=3
    plt.plot(x,ACC,linewidth=width_c,marker='|',color='b',linestyle='dashed',label='BA')
    plt.plot(x,ASR,linewidth=width_c,marker='D',color='r',linestyle='solid',label='ASR')

    plt.ylim(0.0, 1.1)
    # plt.xlim(0,50)
    # plt.xticks([0,10,20,30,40,50])
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend( prop = {'size':18})
    plt.subplots_adjust(left=0.115, right=0.95, top=0.95, bottom=0.165)
    # plt.show()
    if exp is None:
        plt.savefig(f'fig/{mode}_{model}_{dataset}.pdf')
    else:
        plt.savefig(f'fig/{mode}_{exp}_{model}_{dataset}.pdf')
```

```python
def read_txt(file_dir):
    #read txt method
    data = []
    f = open(file_dir)
    for line in f:
        data.append(line)
#     line = f.readline()
#     while line:
#         line = f.readline()
    f.close()
    return data

def str2array(aim_str, split = ", "):
    data = aim_str.split(split)
    for i in range(len(data)):
        data[i] = float(data[i])
    return np.array(data)

def outlier_detection(l1_norm_list, idx_mapping=None):
    print("-" * 30)
    print("Determining whether model is backdoor")
    consistency_constant = 1.4826
    median = np.median(l1_norm_list)
    mad = consistency_constant * np.median(np.abs(l1_norm_list - median))
    min_mad = np.abs(np.min(l1_norm_list) - median) / mad

    print("Median: {}, MAD: {}".format(median, mad))
    print("Anomaly index: {}".format(min_mad))

    if min_mad < 2:
        print("Not a backdoor model")
    else:
        print("This is a backdoor model")
def get_nc(mode, cost='1'):
    index_all = []
    norm_list_all = []
    for i in range(5):
        nc_result = read_txt(f'E:\\work\\datafree_atk\\defends\\neural_cleanse\\r_random_cost{cost}\\results_{mode}{i}\\mnist\\all2one\\all2one_mnist_output.txt')
        index = str2array(nc_result[-2][:-1])
        norm_list = str2array(nc_result[-1][:-1])
        if mode == 'h':
            _ = norm_list[0]
            norm_list[0] = norm_list[2]
            norm_list[2] = _
        index_all.append(index)
        norm_list_all.append(norm_list)
    return np.array(index_all), np.array(norm_list_all)
def draw_norm_bar(norm_list):
    for i in range(5):
        plt.figure()
        x = range(10)
        y = norm_list[i]
        color = ['b']*10
        color[2] = 'r'
        plt.bar(x, y, color=color, width=0.5)
        plt.show()
```
